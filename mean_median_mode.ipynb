```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en'])
```

```python
v=input()
Image(v)
```

```python
output = reader.readtext('I.jpg')
```

```python
output
```

```python
cord = output[0][0]
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
```

```python
image = cv2.imread('I.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
cord_1= output[0][1]
def split(word):
    return [char for char in word]
                  
```

```python
cord_2=split(cord_1)
```

```python
cord_1=int(cord_1)
```

```python
cord_1
```

```python
cord_2
```

```python
for i in range(0, len(cord_2)): 
    cord_2[i] = int(cord_2[i])
```

```python
cord_2
```

```python
def mean(lst): 
    return sum(lst) / len(lst) 
```

```python
a=mean(cord_2)
a=int(a)
a=str(a)
print (a)
```

```python
import statistics
c=statistics.mode(cord_2)
c=str(c)
```

```python
cord_2.sort() 
d = len(cord_2) // 2
e = (cord_2[d] + cord_2[~d]) / 2
e=int(e)
e=str(e)
```

```python
import numpy as np

image = cv2.imread('I.jpg',cv2.IMREAD_UNCHANGED)
font = cv2.FONT_HERSHEY_SIMPLEX 
position = (220,80)
position1 = (100,80)
position11 = (309,81)
position12 = (189,81)
position2 = (3,81)
position22 = (70,81)
fontScale = 1
color = (255, 0, 0)  
thickness = 1 
image = cv2.putText(image, "mean", position, font,  fontScale, color, thickness, cv2.LINE_AA) 
image = cv2.putText(image, "mode", position1, font,  fontScale, color, thickness, cv2.LINE_AA)
image = cv2.putText(image, a, position11, font,  fontScale, color, thickness, cv2.LINE_AA) 
image = cv2.putText(image, c, position12, font,  fontScale, color, thickness, cv2.LINE_AA)
image = cv2.putText(image,"med", position2, font,  fontScale, color, thickness, cv2.LINE_AA) 
image = cv2.putText(image, e, position22, font,  fontScale, color, thickness, cv2.LINE_AA)




cv2.imshow('outputimage', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
