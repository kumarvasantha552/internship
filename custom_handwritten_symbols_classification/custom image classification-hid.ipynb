```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pickle
```

```python
pickle_in = open("X_train","rb")
x_train= pickle.load(pickle_in)
```

```python
pickle_in = open("X_Data","rb")
x_test= pickle.load(pickle_in)
```

```python
pickle_in = open("Y_train","rb")
y_train = pickle.load(pickle_in)
```

```python
pickle_in = open("Y_Data","rb")
y_test = pickle.load(pickle_in)
```

```python
print (x_train[0].shape)
```

```python
plt.matshow(x_train[0])
```

```python
y_train[0]
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
x_train[0]
```

```python
x_train_flattened = x_train.reshape(len(x_train), 28*28)
x_test_flattened = x_test.reshape(len(x_test), 28*28)
```

```python
x_train_flattened
```

```python
x_train_flattened.shape
```

```python
model_hid = keras.Sequential([
    keras.layers.Dense(27, input_shape=(784,), activation='relu'),
    keras.layers.Dense(3, activation='sigmoid')
])

model_hid.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

a=model_hid.fit(x_train_flattened, y_train, epochs=5000)
```

```python
model_hid.evaluate(x_test_flattened, y_test)
```

```python
y_predicted = model_hid.predict(x_test_flattened)
y_predicted[0]
```

```python
plt.matshow(x_test[444])
```

```python
np.argmax(y_predicted[444])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
print(a.history.keys())
```

```python

# summarize history for accuracy
plt.plot(a.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(a.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```
